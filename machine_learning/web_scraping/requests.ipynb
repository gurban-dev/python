requests.get() fetches the raw HTML source code of the page.

Appropriate for scraping static websites where the content is

already present in the HTML source code.
<br><br><br>
driver.get() loads the page in a real browser instance essentially simulating

an end user visiting the page.

```python
url = 'https://hdr.undp.org/data-center/human-development-index#/indicies/HDI'

# driver = webdriver.Chrome()

try:
    '''
    Web scraping starts with sending HTTP requests, such as POST or
    GET, to a website’s server, which returns a response containing
    the needed data.

    GET gets a resource or resources from the server.
    POST creates a resource on the server.
    
    requests returns a Response object, so "response" is the Response
    object which means that it is an instance of the Response class.'''
    response = requests.get(url)

    # Navigate to the HDI page
    # driver.get(url)

    # Response.raise_for_status() will raise an HTTPError if
    # the HTTP request returned an unsuccessful status code.
    response.raise_for_status()

    # Decode the JSON response body (if any) as a Python object.
    data = response.json()

    print(data)

# All exceptions that Requests explicitly raises inherit
# from requests.exceptions.RequestException.
except requests.exceptions.RequestException as err:
    print(f"An error occurred:\n\n{err}")
```

The limitation of the requests library is that it doesn't

parse the extracted HTML data.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Send a request to the webpage
url = "https://hdr.undp.org/data-center/human-development-index#/indicies/HDI"
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    print("Page successfully fetched!")
else:
    print("Failed to retrieve the page.")
    exit()

# Step 2: Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# The table is inside the div with the class 'views-table'
table = soup.find('table', {'class': 'views-table'})

# Step 3: Extract the table headers
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

# Step 4: Extract table rows
rows = []
for tr in table.find_all('tr')[1:]:  # skip header row
    row = [td.text.strip() for td in tr.find_all('td')]
    rows.append(row)

# Step 5: Create a pandas DataFrame
df = pd.DataFrame(rows, columns=headers)

# Step 6: Print or save the DataFrame
print(df.head())  # Print the first few rows to check
# Save to CSV file
df.to_csv('human_development_index.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Set up the Chrome WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Navigate to the HDI data webpage
url = 'https://hdr.undp.org/data-center/human-development-index#/indicies/HDI'
driver.get(url)

# Allow time for the page to load and for the table to populate.  Adjust as needed.
time.sleep(10)

try:
    # Locate the table containing HDI data using a more specific CSS selector.
    # The actual selector will need to be determined by inspecting the page source.
    # This is just an example; replace with the correct selector!
    table = driver.find_element(By.CSS_SELECTOR, 'div.view-indicators__content table')


    # Extract rows from the table
    rows = table.find_elements(By.TAG_NAME, 'tr')

    # Prepare a list to hold HDI data
    hdi_data = []

    # Iterate through each row and extract country names and HDI values
    # Adapt the column indexes based on the actual table structure
    for row in rows[1:]:  # Skip header row
        columns = row.find_elements(By.TAG_NAME, 'td')
        if len(columns) > 2:  # Ensure there are enough columns
            country = columns[0].text.strip()  # Country name is in the first column
            hdi_value = columns[2].text.strip()  # HDI value is in the third column (example)

            # Handle potential errors if the HDI value is not a number
            try:
                hdi_value = float(hdi_value)  # Try converting to a float
                hdi_data.append({'Country': country, 'HDI': hdi_value})
            except ValueError:
                print(f"Skipping row for {country} due to invalid HDI value: {hdi_value}")


    # Convert the list of dictionaries into a DataFrame for easy handling
    hdi_df = pd.DataFrame(hdi_data)

    # Display the DataFrame or save it to a CSV file
    print(hdi_df)
    hdi_df.to_csv('hdi_data.csv', index=False)


except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser after scraping is complete, even if errors occur
    driver.quit()
```
