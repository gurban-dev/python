```python
import requests
from lxml import html

url = 'https://oxylabs.io/blog'

# Send an HTTP GET request to the above URL and
# get back an object of the Response class.
response = requests.get(url)

print(f'response.text: {response.text}')

# Pass the HTML content from the response to convert it to
# a lxml.html.HtmlElement object, which is a tree-like
# representation of the HTML document.
tree = html.fromstring(response.text)
```

The object called "tree" can now be invoked to extract the required data.

```python
tree
```

The contains() function selects `<a>` elements that have a <br>
class attribute containing the substring "e1uzm1ya1".

```python
blog_titles = tree.xpath('//a[contains(@class, "e1uzm1ya1")]')
for title in blog_titles:
    print(title.text)
```
