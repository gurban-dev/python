To create a new excel spreadsheet:

```python
'''
ExcelWriter is a class in the Pandas library that allows you to write
DataFrame objects to an Excel file.

The first argument is the name of the excel spreadsheet being created.
"engine" is the engine that will be used for writing. It is an optional
parameter, so it has a default value if not assigned a value.
'''
writer = pd.ExcelWriter('Splitted Raw Data.xlsx', engine = 'xlsxwriter')

for state in df['Location'].unique():
    print(f"df[\'Location\'] == state:\n{df['Location'] == state} \
        \n\ndf[df[\'Location\'] == state]:\n{df[df['Location'] == state]}")

    # Retrieve all of the rows in the pandas dataframe called "df"
    # where the value in the "Location" column is equal to state.
    new_df = df.loc[df['Location'] == state]

    # The "sheet_name" parameter is assigned the name of sheet which
    # will contain DataFrame which is "state" on this case.

    # "index" is set to false because indexes don't need to be written
    # for the rows.
    new_df.to_excel(excel_writer = writer, sheet_name = state, index = False)

# Call close() to save and close any opened file handles.
writer.close()
```
